```python
#busca caracteres rraros
with open("pronostico-ventas.csv", "rb") as f:
    contenido = f.read()

# Tomamos un margen antes y después del byte 684 para ver el contexto
inicio = max(0, 684 - 20)
fin = min(len(contenido), 684 + 20)

print("Bytes problemáticos:", contenido[684 : 684 + 5])
print("\nContexto en texto (reemplazando lo que no se entiende):")
print(contenido[inicio:fin].decode("utf-8", errors="replace"))
```
